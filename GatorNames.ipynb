```python
# import libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
url = "http://dos.sunbiz.org/scripts/ficregl.exe?action=REGPRE&docnum=G14000071573"
```

```python
data_list = []
date_list = []
```

```python
count = 1
while (count < 2):
    page = urlopen(url)
    soup = BeautifulSoup(page, "html.parser") # parse the html using beautiful soup and store in variable `soup`
    divs = soup.findAll("table", {"summary": "This table displays the name lists for the 6 column headings."}) # Take out the <div> of otherInfo and get its value
    x = len(data_list)
    for div in divs:
        rows = div.findAll('tr')
        for row in rows[1:]:
            data_list.append(row.findAll('td'))
    for j in range(x, len(data_list)):
        for k in range(0, len(data_list[j])):
            data_list[j][k] = data_list[j][k].text.strip()
            
    urls = soup.findAll('a')
    for i in range(2, len(urls)-4):
        link = "http://dos.sunbiz.org" + urls[i]['href']
        date_page = urlopen(link)
        small_soup = BeautifulSoup(date_page, "html.parser")
        table = small_soup.find('table', attrs={'summary': 'This table contains filing information.'})
        td_list = table.findAll('td')
        date_list.append([td_list[6].text.strip()])
    
    temp = [div.find('a') for div in soup.find('div', attrs={'class':'search-nav bottom'})]
    end = temp[1].find_next('a')
    url = "http://dos.sunbiz.org" + end['href']
    if "gator" not in data_list[-1][0].lower():
        count = 2
```

```python
final_list = []
```

```python
for i in range (0, len(data_list)):
    final_list.append(data_list[i] + date_list[i])
```

```python
# open a csv file with append, so old data will not be erased
with open('GatorNames&Dates.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(["name", "address", "city", "state", "county", "status", "date"]) # write csv file headers
    for i in range (0, len(data_list)):
        writer.writerow(final_list[i])
```
